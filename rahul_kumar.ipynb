```python
#Lets create the function to create and dispaly the displot and boxplot for the columns
def createPlotInSameDataFrame(df, column):
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    
    # Distribution plot
    sns.histplot(df.loc[df[column].notnull(), column], kde=True, ax=axes[0])
    axes[0].set_title('Distribution of ' + column_titles_detail[column])
    
    # Box plot
    sns.boxplot(x=df[column], ax=axes[1])
    axes[1].set_title('Boxplot of ' + column_titles_detail[column])
    
    plt.tight_layout()
    plt.show()
```

```python
# create  function to create the boxplot for multiple columns in a single plot
def createBoxPlot(df, columns):
    fig, p = plt.subplots(1, len(columns), figsize=(16, 4))
    for i, column in enumerate(columns):
        sns.boxplot(x=column, data=df, ax=p[i])
        p[i].set_xlabel(column_titles_detail[column])
    plt.tight_layout()
    plt.show()
```

```python
#using createBoxPlot function to create the boxplot for multiple columns
createBoxPlot(final_loan, ['loan_amnt', 'annual_inc', 'int_rate', 'dti', 'installment'])
```

```python
#as we can see there are outliers in the columns, we can remove the outliers
#remove outliers


# Define a function to remove outliers based on IQR
def removeOutliersBasedOnIqr(df, column, threshold):
    for column in columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (threshold * iqr)
        upper_bound = q3 + (threshold * iqr)
        df = df[(df[column] > lower_bound) & (df[column] < upper_bound)]
    return df

# Remove outliers from the columns
# Define the columns to remove outliers
columns = ['loan_amnt', 'annual_inc', 'int_rate', 'dti', 'installment']
final_loan = removeOutliersBasedOnIqr(final_loan, columns, 1.5)  
    
    
```

```python
#using createBoxPlot function to create the boxplot for multiple columns
createBoxPlot(final_loan, ['loan_amnt', 'annual_inc', 'int_rate', 'dti', 'installment'])
```

```python
#Now lets try to find the correlation between the columns. 
#Under the loan status coloumn we have 3 values, Fully Paid, Charged Off, Default.
#Now lets try to find the correlation between the loan_status (Charged Off) and other columns
# Will try to extract the columns which are highly correlated with loan_status (Charged Off)
# will compare the charged off proportion with Annual Income, Interest Rate, Home Ownership, Loan Purpose, DTI, 
#pub_rec_bankruptcies, State, Grade.
```

```python
def plot_charged_off_proportion(df, column_to_compare, bins):
    # Create bins for the comparison column
    df['comparison_bin'] = pd.cut(df[column_to_compare], bins=bins, include_lowest=True)
    
    # Calculate the proportion of 'Charged Off' loans within each bin
    charged_off_proportion = df[df['loan_status'] == 'Charged Off'].groupby('comparison_bin').size() / df.groupby('comparison_bin', observed=True).size()
    
    # Reset index to turn the Series into a DataFrame
    charged_off_proportion = charged_off_proportion.reset_index(name='proportion_charged_off')
    
    # Create a figure with two subplots
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Bar plot
    sns.barplot(x='comparison_bin', y='proportion_charged_off', data=charged_off_proportion, ax=axes[0])
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)
    axes[0].set_title(f'Proportion of Charged Off Loans by {column_to_compare} Bins')
    axes[0].set_xlabel(f'{column_to_compare} Bins')
    axes[0].set_ylabel('Proportion of Charged Off Loans')
    
    # Annotate the bars with the proportion values
    for p in axes[0].patches:
        axes[0].annotate(format(p.get_height(), '.2f'), 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha = 'center', va = 'center', 
                         xytext = (0, 9), 
                         textcoords = 'offset points')
    
    # Box plot
    sns.boxplot(x='loan_status', y=column_to_compare, data=df[df['loan_status'] == 'Charged Off'], ax=axes[1])
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
    axes[1].set_title(f'{column_to_compare} BoxPlot for Charged Off Loans')
    axes[1].set_xlabel(f'{column_to_compare}')
    axes[1].set_ylabel(column_to_compare)
    
    plt.tight_layout()
    plt.show()
```

```python
#can we make a code changes to give colours to the bars in the bar plot

def plot_charged_off_proportion_categorical(df, column_to_compare):
    # Check if the necessary columns exist in the DataFrame
    if 'loan_status' not in df.columns:
        raise KeyError("The DataFrame does not contain the 'loan_status' column.")
    if column_to_compare not in df.columns:
        raise KeyError(f"The DataFrame does not contain the '{column_to_compare}' column.")
    
    # Calculate the proportion of 'Charged Off' loans within each category
    charged_off_proportion = df[df['loan_status'] == 'Charged Off'].groupby(column_to_compare).size() / df.groupby(column_to_compare, observed=True).size()
    
    # Reset index to turn the Series into a DataFrame
    charged_off_proportion = charged_off_proportion.reset_index(name='proportion_charged_off')
    
    # Create a bar plot
    plt.figure(figsize=(16, 6))
    ax = sns.barplot(x=column_to_compare, y='proportion_charged_off', data=charged_off_proportion, palette='viridis')
    
    # Add numbers on top of bars
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    
    plt.xticks(rotation=45)
    plt.title(f'Proportion of Charged Off Loans by {column_to_compare}')
    plt.xlabel(column_to_compare)
    plt.ylabel('Proportion of Charged Off Loans')
    plt.tight_layout()
    plt.show()
```
